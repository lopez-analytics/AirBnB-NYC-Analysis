## 04. Clean listing.csv
### Check for duplicates, mixed data types, and explore missing values

```python
# Check for duplicate rows in listing.csv
duplicates = listing_csv[listing_csv.duplicated()]

# Display the duplicate rows
print(duplicates)
```

```python
# Check for mixed data types in listing.csv
for col in listing_csv.columns.tolist():
  weird = (listing_csv[[col]].applymap(type) != listing_csv[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (listing_csv[weird]) > 0:
    print (col)
```

```python
# Explore returned columns for mixed data types
columns_to_check = ['name','host_name','last_review', 'license']  

# Loop through each specified column
for col in columns_to_check:
    # Get the types of each entry in the column
    column_types = listing_csv[col].apply(type)
    
    # Filter rows with mixed types by comparing each entry's type to the type of the first entry
    mixed_data = listing_csv[column_types != column_types.iloc[0]]
    
    # Display the mixed data for the current column if any mixed types are found
    if not mixed_data.empty:
        print(f"Mixed data in column '{col}':")
        # Print the unique mixed data values in a compact format
        print(mixed_data[col].unique())  # Print unique values for readability
```

```python
# Specify the column you want to inspect (e.g., 'name')
column_to_inspect = 'name'

# Find rows where the data type is 'float' in the specified column
float_rows =listing_csv[listing_csv[column_to_inspect].apply(lambda x: isinstance(x, float))]

# Display the rows with float data types
print(f"Rows in column '{column_to_inspect}' with float data types:")
print(float_rows)
```

```python
# Drop rows since they are also both missing prices
 
# Find two rows to drop
rows_to_drop = listing_csv[listing_csv[column_to_inspect].apply(lambda x: isinstance(x, float))].index

# Drop the rows with float data types
listing_csv_cleaned = listing_csv.drop(rows_to_drop)

# Optionally reset the index if you want a clean index after dropping
listing_csv_cleaned = listing_csv_cleaned.reset_index(drop=True)

# Check output
listing_csv_cleaned.head()
```

```python
# Find the number of missing values in each column
missing_values = listing_csv_cleaned.isnull().sum()

# Filter out columns with no missing values
missing_values = missing_values[missing_values > 0]

# Print the columns with missing values and the count of missing values
print("Columns with missing values:")
print(missing_values)

# Optionally, you can also print the percentage of missing values
missing_percentage = (listing_csv_cleaned.isnull().mean() * 100).round(2)
missing_percentage = missing_percentage[missing_percentage > 0]

print("\nPercentage of missing values in each column:")
print(missing_percentage)
```

```python
# Drop the 'license' column from the DataFrame
listing_csv_cleaned = listing_csv_cleaned.drop(columns=['license'])

# Display the updated DataFrame to confirm the column has been dropped
listing_csv_cleaned.head()
```

### Address missing values in 'prices'
#### There are 38.98% of prices are missing from listings. As this is a key variable, the missing variables will be input using the mean price of the listing's respective neighborhood. 

```python
# Calculate the mean price for each neighborhood
neighborhood_avg_prices = listing_csv_cleaned.groupby('neighbourhood')['price'].mean()

# Replace missing prices with the average price of the respective neighborhood
listing_csv_cleaned['price'] = listing_csv_cleaned.apply(
    lambda row: neighborhood_avg_prices[row['neighbourhood']] if pd.isnull(row['price']) else row['price'],
    axis=1
)

# Display the first few rows to confirm the changes
listing_csv_cleaned.head()
```

```python
# Check for missing values in the 'price' column
missing_price = listing_csv_cleaned['price'].isnull().sum()

# Display the result
print(f"Number of missing values in 'price' column: {missing_price}")
```

```python
# Filter to show rows where 'price' is NaN
missing_price_rows = listing_csv_cleaned[listing_csv_cleaned['price'].isnull()]

# Display the rows with missing values in 'price'
print(missing_price_rows)
```

```python
# Calculate the average price for properties in 'New Dorp'
average_price_new_dorp = listing_csv_cleaned[listing_csv_cleaned['neighbourhood'] == 'New Dorp']['price'].mean()

# Replace the missing value in 'price' with the calculated average price
listing_csv_cleaned['price'] = listing_csv_cleaned['price'].fillna(average_price_new_dorp)

# Check output
print(listing_csv_cleaned[listing_csv_cleaned['neighbourhood'] == 'New Dorp'].head())
```

```python
# Filter to show all listings in 'New Dorp'
new_dorp_listings = listing_csv_cleaned[listing_csv_cleaned['neighbourhood'] == 'New Dorp']

# Display the listings in New Dorp
print(new_dorp_listings)
```

```python
# Drop rows where the neighbourhood is 'New Dorp'
listing_csv_cleaned = listing_csv_cleaned[listing_csv_cleaned['neighbourhood'] != 'New Dorp']

# Check output
listing_csv_cleaned.head()
```

```python
# Round the 'price' col to 2 decimal points
listing_csv_cleaned['price'] = listing_csv_cleaned['price'].round(2)

# Check output
listing_csv_cleaned[['price']].head()
```

```python
# Explore missing reiviews rows
# Filter to show rows where 'reviews_per_month' or 'last_review' is NaN
missing_reviews_rows = listing_csv_cleaned[listing_csv_cleaned['reviews_per_month'].isnull() | listing_csv_cleaned['last_review'].isnull()]

# Display the rows with missing values in 'reviews_per_month' and 'last_review'
missing_reviews_rows
```

```python
# Filter to show rows where 'reviews_per_month' or 'last_review' is NaN
missing_reviews_rows = listing_csv_cleaned[listing_csv_cleaned['reviews_per_month'].isnull() | listing_csv_cleaned['last_review'].isnull()]

# Display up to 100 rows with missing values in 'reviews_per_month' and 'last_review'
missing_reviews_rows.head(100)
```
